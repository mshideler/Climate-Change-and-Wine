```python
# Import dependency
import pandas as pd
```

```python
# Read projected precipitation data in CSV to a DataFrame
file_path1 = "../Raw Data/Italy Precipitation Predict.csv"
precip_df = pd.read_csv(file_path1)
precip_df.head()
```

```python
# Rename unnamed column
precip_df = precip_df.rename(columns={'Unnamed: 0': 'Year'})
precip_df.head()
```

```python
# Create new dataframe with applicable provinces
new_precip_df = precip_df[['Year', 'Piemonte', 'Sardegna', 'Sicilia', 'Toscana', 'Veneto']]

# Rename columns to ID them as precipitation columns
new_precip_df = new_precip_df.rename(columns={'Piemonte': 'Piemonte Precip', 'Sardegna': 'Sardegna Precip', 'Sicilia': 'Sicilia Precip', 'Toscana': 'Toscana Precip', 'Veneto': 'Veneto Precip'})

new_precip_df.head()
```

```python
# Read projected temperature data in CSV to a DataFrame
file_path2 = "../Raw Data/Italy Temp Predict.csv"
temp_df = pd.read_csv(file_path2)
temp_df.head()
```

```python
# Rename unnamed column
temp_df = temp_df.rename(columns={'Unnamed: 0': 'Year'})
temp_df.head()
```

```python
# Create new dataframe with applicable provinces
new_temp_df = temp_df[['Year', 'Piemonte', 'Sardegna', 'Sicilia', 'Toscana', 'Veneto']]

# Rename columns to ID them as temperature columns
new_temp_df = new_temp_df.rename(columns={'Piemonte': 'Piemonte Temp', 'Sardegna': 'Sardegna Temp', 'Sicilia': 'Sicilia Temp', 'Toscana': 'Toscana Temp', 'Veneto': 'Veneto Temp'})

new_temp_df.head()
```

```python
# Merge the new_precip and new_temp DataFrames into one DataFrame
merge_df = pd.merge(new_precip_df, new_temp_df, on='Year')
merge_df.head()
```

```python
# Create province DataFrame with year, precipitation and temperature
piemonte_df = merge_df[['Year', 'Piemonte Precip', 'Piemonte Temp']]

# Add a column to specify which province the precipitation and temperature data is for
piemonte_df['Province'] = pd.Series(['Piemonte' for x in range(len(piemonte_df.index))])

# Rename the precipitation and temperature columns by replacing the province name with "Projected"
piemonte_df = piemonte_df.rename(columns={'Piemonte Precip': 'Projected Precip', 'Piemonte Temp': 'Projected Temp'})

# Reorder the columns
piemonte_df = piemonte_df[['Province', 'Year', 'Projected Precip', 'Projected Temp']]

piemonte_df.head()
```

```python
# Create province DataFrame with year, precipitation and temperature
sardegna_df = merge_df[['Year', 'Sardegna Precip', 'Sardegna Temp']]

# Add a column to specify which province the precipitation and temperature data is for
sardegna_df['Province'] = pd.Series(['Sardegna' for x in range(len(sardegna_df.index))])

# Rename the precipitation and temperature columns by replacing the province name with "Projected"
sardegna_df = sardegna_df.rename(columns={'Sardegna Precip': 'Projected Precip', 'Sardegna Temp': 'Projected Temp'})

# Reorder the columns
sardegna_df = sardegna_df[['Province', 'Year', 'Projected Precip', 'Projected Temp']]

sardegna_df.head()
```

```python
# Create province DataFrame with year, precipitation and temperature
sicilia_df = merge_df[['Year', 'Sicilia Precip', 'Sicilia Temp']]

# Add a column to specify which province the precipitation and temperature data is for
sicilia_df['Province'] = pd.Series(['Sicilia' for x in range(len(sicilia_df.index))])

# Rename the precipitation and temperature columns by replacing the province name with "Projected"
sicilia_df = sicilia_df.rename(columns={'Sicilia Precip': 'Projected Precip', 'Sicilia Temp': 'Projected Temp'})

# Reorder the columns
sicilia_df = sicilia_df[['Province', 'Year', 'Projected Precip', 'Projected Temp']]

sicilia_df.head()
```

```python
# Create province DataFrame with year, precipitation and temperature
toscana_df = merge_df[['Year', 'Toscana Precip', 'Toscana Temp']]

# Add a column to specify which province the precipitation and temperature data is for
toscana_df['Province'] = pd.Series(['Toscana' for x in range(len(toscana_df.index))])

# Rename the precipitation and temperature columns by replacing the province name with "Projected"
toscana_df = toscana_df.rename(columns={'Toscana Precip': 'Projected Precip', 'Toscana Temp': 'Projected Temp'})

# Reorder the columns
toscana_df = toscana_df[['Province', 'Year', 'Projected Precip', 'Projected Temp']]

toscana_df.head()
```

```python
# Create province DataFrame with year, precipitation and temperature
veneto_df = merge_df[['Year', 'Veneto Precip', 'Veneto Temp']]

# Add a column to specify which province the precipitation and temperature data is for
veneto_df['Province'] = pd.Series(['Veneto' for x in range(len(veneto_df.index))])

# Rename the precipitation and temperature columns by replacing the province name with "Projected"
veneto_df = veneto_df.rename(columns={'Veneto Precip': 'Projected Precip', 'Veneto Temp': 'Projected Temp'})

# Reorder the columns
veneto_df = veneto_df[['Province', 'Year', 'Projected Precip', 'Projected Temp']]

veneto_df.head()
```

```python
# Combine DataFrames into one
concat_df = pd.concat([piemonte_df, sardegna_df, sicilia_df, toscana_df, veneto_df])
concat_df
```

```python
# Save concat_df to CSV
concat_df.to_csv('../Raw Data/projected_weather_italy.csv', index=False)
```
