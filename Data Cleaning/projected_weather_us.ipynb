```python
# Import dependency
import pandas as pd
```

```python
# Read projected precipitation data in CSV to a DataFrame
file_path1 = "../Raw Data/US Precipitation Predict.csv"
precip_df = pd.read_csv(file_path1)
precip_df.head()
```

```python
# Rename unnamed column
precip_df = precip_df.rename(columns={'Unnamed: 0': 'Year'})
precip_df.head()
```

```python
# Create new dataframe with applicable provinces
new_precip_df = precip_df[['Year', 'California', 'New York', 'Oregon', 'Washington']]

# Rename columns to ID them as precipitation columns
new_precip_df = new_precip_df.rename(columns={'California': 'California Precip', 'New York': 'New York Precip', 'Oregon': 'Oregon Precip', 'Washington': 'Washington Precip'})

new_precip_df.head()
```

```python
# Read projected temperature data in CSV to a DataFrame
file_path2 = "../Raw Data/US Temp Predict.csv"
temp_df = pd.read_csv(file_path2)
temp_df.head()
```

```python
# Rename unnamed column
temp_df = temp_df.rename(columns={'Unnamed: 0': 'Year'})
temp_df.head()
```

```python
# Create new dataframe with applicable provinces
new_temp_df = temp_df[['Year', 'California', 'New York', 'Oregon', 'Washington']]

# Rename columns to ID them as temperature columns
new_temp_df = new_temp_df.rename(columns={'California': 'California Temp', 'New York': 'New York Temp', 'Oregon': 'Oregon Temp', 'Washington': 'Washington Temp'})

new_temp_df.head()
```

```python
# Merge the new_precip and new_temp DataFrames into one DataFrame
merge_df = pd.merge(new_precip_df, new_temp_df, on='Year')
merge_df.head()
```

```python
# Create province DataFrame with year, precipitation and temperature
ca_df = merge_df[['Year', 'California Precip', 'California Temp']]

# Add a column to specify which province the precipitation and temperature data is for
ca_df['Province'] = pd.Series(['California' for x in range(len(ca_df.index))])

# Rename the precipitation and temperature columns by replacing the province name with "Projected"
ca_df = ca_df.rename(columns={'California Precip': 'Projected Precip', 'California Temp': 'Projected Temp'})

# Reorder the columns
ca_df = ca_df[['Province', 'Year', 'Projected Precip', 'Projected Temp']]

ca_df.head()
```

```python
# Create province DataFrame with year, precipitation and temperature
ny_df = merge_df[['Year', 'New York Precip', 'New York Temp']]

# Add a column to specify which province the precipitation and temperature data is for
ny_df['Province'] = pd.Series(['New York' for x in range(len(ny_df.index))])

# Rename the precipitation and temperature columns by replacing the province name with "Projected"
ny_df = ny_df.rename(columns={'New York Precip': 'Projected Precip', 'New York Temp': 'Projected Temp'})

# Reorder the columns
ny_df = ny_df[['Province', 'Year', 'Projected Precip', 'Projected Temp']]

ny_df.head()
```

```python
# Create province DataFrame with year, precipitation and temperature
or_df = merge_df[['Year', 'Oregon Precip', 'Oregon Temp']]

# Add a column to specify which province the precipitation and temperature data is for
or_df['Province'] = pd.Series(['Oregon' for x in range(len(or_df.index))])

# Rename the precipitation and temperature columns by replacing the province name with "Projected"
or_df = or_df.rename(columns={'Oregon Precip': 'Projected Precip', 'Oregon Temp': 'Projected Temp'})

# Reorder the columns
or_df = or_df[['Province', 'Year', 'Projected Precip', 'Projected Temp']]

or_df.head()
```

```python
# Create province DataFrame with year, precipitation and temperature
wa_df = merge_df[['Year', 'Washington Precip', 'Washington Temp']]

# Add a column to specify which province the precipitation and temperature data is for
wa_df['Province'] = pd.Series(['Washington' for x in range(len(wa_df.index))])

# Rename the precipitation and temperature columns by replacing the province name with "Projected"
wa_df = wa_df.rename(columns={'Washington Precip': 'Projected Precip', 'Washington Temp': 'Projected Temp'})

# Reorder the columns
wa_df = wa_df[['Province', 'Year', 'Projected Precip', 'Projected Temp']]

wa_df.head()
```

```python
# Combine DataFrames into one
concat_df = pd.concat([ca_df, ny_df, or_df, wa_df])
concat_df
```

```python
# Save concat_df to CSV
concat_df.to_csv('../Raw Data/projected_weather_us.csv', index=False)
```
